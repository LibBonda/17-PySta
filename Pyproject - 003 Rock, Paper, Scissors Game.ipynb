```python
import random, sys

print('ROCK, PAPER, SCISSORS')

wins = 0
losses = 0
ties = 0

options = ['Rock', 'Paper', 'Scissors']

while True:
    # Calculate total games played
    total_games = wins + losses + ties

    # Display the current score with percentages
    if total_games > 0:
        win_percentage = (wins / total_games) * 100
        loss_percentage = (losses / total_games) * 100
        tie_percentage = (ties / total_games) * 100
        print('%s Wins (%.0f%%), %s Losses (%.0f%%), %s Ties (%.0f%%)' % 
              (wins, win_percentage, losses, loss_percentage, ties, tie_percentage))
    else:
        print('%s Wins, %s Losses, %s Ties' % (wins, losses, ties))

    # Prompt the player for input
    playerMove = input('Enter your move: (r)ock (p)aper (s)cissors or (q)uit: ').lower()

    # Handle quitting the game
    if playerMove == 'q':
        print('Thanks for playing!')
        break  # Breaks the loop and ends the game. Note break is used in IDEs instead of 'sys.exit()' in gaming systems

    # Validate the player's move
    if playerMove not in ['r', 'p', 's']:
        print("Invalid input. Please enter 'r', 'p', 's', or 'q'.")
        continue

    # Convert player's move to full word for display
    if playerMove == 'r':
        print('You chose ROCK.')
    elif playerMove == 'p':
        print('You chose PAPER.')
    elif playerMove == 's'
        print('You chose SCISSORS.')

    # Computer makes a random choice
    computerChoice = random.choice(options)
    print('Computer chose ' + computerChoice.upper() + '.')

    # Display the full "Player vs Computer" choices
    if playerMove == 'r':
        print('ROCK vs ' + computerChoice.upper())
    elif playerMove == 'p':
        print('PAPER vs ' + computerChoice.upper())
    elif playerMove == 's':
        print('SCISSORS vs ' + computerChoice.upper())

    # Determine the computer's corresponding move letter
    if computerChoice == 'Rock':
        computerMove = 'r'
    elif computerChoice == 'Paper':
        computerMove = 'p'
    elif computerChoice == 'Scissors':
        computerMove = 's'

    # Determine the result of the game (win/loss/tie)
    if playerMove == computerMove:
        print("It's a TIE!")
        ties += 1
    elif (playerMove == 'r' and computerMove == 's') or \
         (playerMove == 'p' and computerMove == 'r') or \
         (playerMove == 's' and computerMove == 'p'):
        print("You WIN!")
        wins += 1
    else:
        print("You LOSE!")
        losses += 1
```

### Rock Paper Scissors Option 3

```python
import random

# Variables to keep track of wins, losses, and ties
wins = 0
losses = 0
ties = 0

# List of possible outcomes
outcome = ["Rock", "Paper", "Scissors"]

while True:
    try:
        userOption = int(input('''\n
    Game Start
    1 Yes
    2 No | Exit
    '''))

        # Check if the user wants to start the game or exit
        if userOption == 1:
            # Play until the user wants to stop
            while True:
                try:
                    userInput = int(input('''\n
                Choose:
                1 Rock
                2 Paper
                3 Scissors
                4 Exit Game
                '''))

                    # User chooses to exit the game
                    if userInput == 4:
                        print(f"\nTotal Results: Wins: {wins}, Losses: {losses}, Ties: {ties}")
                        break

                    # Map user input to the correct choice
                    if userInput == 1:
                        userChoice = "Rock"
                    elif userInput == 2:
                        userChoice = "Paper"
                    elif userInput == 3:
                        userChoice = "Scissors"
                    else:
                        print("Invalid input, please choose a valid option (1, 2, 3, or 4).")
                        continue

                    # Computer randomly chooses from the outcome list
                    computerChoice = random.choice(outcome)
                    print(f"\nUser Choice: {userChoice}")
                    print(f"Computer Choice: {computerChoice}")
                    print()

                    # Determine the winner
                    if userChoice == computerChoice:
                        print("It's a tie!")
                        ties += 1
                    elif (userChoice == "Rock" and computerChoice == "Scissors") or \
                         (userChoice == "Paper" and computerChoice == "Rock") or \
                         (userChoice == "Scissors" and computerChoice == "Paper"):
                        print("You win!")
                        wins += 1
                    else:
                        print("Computer wins!")
                        losses += 1

                except ValueError:
                    print("Invalid input. Please enter a number (1, 2, 3, or 4).")
        elif userOption == 2:
            print("Exiting the game. Goodbye!")
            print(f"\nFinal Results: Wins: {wins}, Losses: {losses}, Ties: {ties}")
            break
        else:
            print("Invalid option. Please choose 1 to play or 2 to exit.")
    except ValueError:
        print("Invalid input. Please enter a number (1 or 2).")
```
