```python
#Rock Paper Scissors
import random, sys

print('ROCK, PAPER, SCISSORS')

#the variable below keep track of the number of wins, losses and ties.Note the count tally starts at 0
wins = 0
losses = 0
ties = 0

while True: # this is the main game loop
    print('%s Wins, %s Losses, %s Ties' % (wins, losses, ties))
    
    while True: #this is the player input loop
        print('Enter your move: (r)ock (p)aper (s)cissors or (q)uit')
        
        playerMove = input()
        
        if playerMove == 'q':
            sys.exit() #Quits the program
            
        if playerMove == 'r' or playerMove == 'p' or playerMove == 's':
            break # break out of player input loop
            print("Type r, p, s or q.")

        #Display players choice
        if playerMove == 'r':
            print('ROCK vs ...')
        elif playerMove == 'p':
            print('PAPER vs...')
        elif playerMove == 's':
            print('SCISSORS vs...')

        # Display computer choice
        randomNum = random.randint(1, 3)
        if randomNum == 1:
            computerMove = 'r'
            print('ROCK')
        elif randomNum == 2:
            computerMove = 'p'
            print('PAPER')
        elif randomNum == 3:
            computerMove = 's'
            print('SCISSORS')
            
        # Display and record the win/loss/tie
        if playerMove == computerMove:
            print("It's a TIE!")
            ties = ties + 1
            
        elif playerMove == 'r' and computerMove == 's':
            print("You WIN!")
            wins = wins + 1
            
        elif playerMove == 'p' and computerMove == 'r':
            print("You WIN!")
            wins = wins + 1
            
        elif playerMove == 's' and computerMove == 'p':
            print("You WIN!")
            wins = wins + 1
            
        elif playerMove == 'r' and computerMove == 'p':
            print("You LOSE!")
            losses = losses + 1

        elif playerMove == 'p' and computerMove == 's':
            print("You LOSE!")
            losses = losses + 1

        elif playerMove == 's' and computerMove == 'r':
            print("You LOSE!")
            losses = losses + 1
        
    
```

```python
import random, sys

print('ROCK, PAPER, SCISSORS')

wins = 0
losses = 0
ties = 0

options = ['Rock', 'Paper', 'Scissors']

while True:
    # Calculate total games played
    total_games = wins + losses + ties

    # Display the current score with percentages
    if total_games > 0:
        win_percentage = (wins / total_games) * 100
        loss_percentage = (losses / total_games) * 100
        tie_percentage = (ties / total_games) * 100
        print('%s Wins (%.0f%%), %s Losses (%.0f%%), %s Ties (%.0f%%)' % 
              (wins, win_percentage, losses, loss_percentage, ties, tie_percentage))
    else:
        print('%s Wins, %s Losses, %s Ties' % (wins, losses, ties))

    # Prompt the player for input
    playerMove = input('Enter your move: (r)ock (p)aper (s)cissors or (q)uit: ').lower()

    # Handle quitting the game
    if playerMove == 'q':
        print('Thanks for playing!')
        break  # Breaks the loop and ends the game. Note break is used in IDEs instead of 'sys.exit()' in gaming systems

    # Validate the player's move
    if playerMove not in ['r', 'p', 's']:
        print("Invalid input. Please enter 'r', 'p', 's', or 'q'.")
        continue

    # Convert player's move to full word for display
    if playerMove == 'r':
        print('You chose ROCK.')
    elif playerMove == 'p':
        print('You chose PAPER.')
    elif playerMove == 's'
        print('You chose SCISSORS.')

    # Computer makes a random choice
    computerChoice = random.choice(options)
    print('Computer chose ' + computerChoice.upper() + '.')

    # Display the full "Player vs Computer" choices
    if playerMove == 'r':
        print('ROCK vs ' + computerChoice.upper())
    elif playerMove == 'p':
        print('PAPER vs ' + computerChoice.upper())
    elif playerMove == 's':
        print('SCISSORS vs ' + computerChoice.upper())

    # Determine the computer's corresponding move letter
    if computerChoice == 'Rock':
        computerMove = 'r'
    elif computerChoice == 'Paper':
        computerMove = 'p'
    elif computerChoice == 'Scissors':
        computerMove = 's'

    # Determine the result of the game (win/loss/tie)
    if playerMove == computerMove:
        print("It's a TIE!")
        ties += 1
    elif (playerMove == 'r' and computerMove == 's') or \
         (playerMove == 'p' and computerMove == 'r') or \
         (playerMove == 's' and computerMove == 'p'):
        print("You WIN!")
        wins += 1
    else:
        print("You LOSE!")
        losses += 1
```
