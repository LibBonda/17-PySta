#### Matplotlib Plotting

##### Plotting x and y points
- The plot() function is used to draw points(markers) in a diagram.
- By default, the plot() function draws a line from point to point.
- The function takes parameters for specifying points in the diagram.
- Parameter 1 is an array containing the points on the x-axis.
- Parameter 2 is an array containing the points on the y-axis.

```python
# Draw a line in a diagram from position (1, 3) to position (8, 10):
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

##### Plotting Without Line
- To plot only the markers, use shortcut string notation parameter 'o', which means 'rings'.

```python
#Draw two points in the diagram, one at position (1, 3) and one in position (8, 10):
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

##### Multiple Points
- It's possible to plot as many points as possible, just make sure you have the same number of points in both axis.

```python
#Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

##### Default X-Points
If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

```python
#Plotting without x-points:
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 9])

plt.plot(ypoints)
plt.show()
```

#### Matplotlib Markers

##### Markers
- Use keyword marker to emphasise each point with a specified marker

```python
#Mark each point with a circle:
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([1, 2, 3, 7, 12])

plt.plot(ypoints, marker = 'o')
plt.show()
```

#### Format Strings fmt
- You can also use the shortcut string notation parameter to specify the marker.
- This parameter is also called fmt, and is written with this syntax:

marker|line|color

```python
#Mark each point with a circle: use fmt marker|line|color
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
```

#### Marker Size
- You can use the keyword argument markersize or the shorter version, ms to set the size of the markers:

```python
#Set the size of the markers to 20:
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 7, 10])

plt.plot(ypoints, marker = 'o', ms =15)
plt.show()
```

#### Marker Color
- Use keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:

```python
#Set the EDGE color to red:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([1, 2, 5, 9, 15, 25])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
```

#### makrkerfacecolor
- Use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers:

```python
#Set the FACE color to red:
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

```python
#Use both the mec and mfc arguments to color the entire marker:
#Set the color of both the edge and the face to red:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
```

##### Use Hexadecimal color values, or any of the 140 supported color names.:
- Mark each point with a beautiful green color:

```python
#plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o-.g', ms = 20, mec = '#4CAF50', mfc = 'hotpink')
plt.show()
```

#### Matplotlib Line
- Use the keyword argument linestyle, or shorter ls, to change the style of the plotted line:

##### Shorter Syntax
- The line style can be written in a shorter syntax:
- linestyle can be written as ls.
- dotted can be written as :.
- dashed can be written as --.

```python
#Use a dotted line:
import matplotlib.pyplot as plt
import numpy as np


ypoints = np.array([2, 7, 0, 9])
ypoints2 = np.array([3, 8, 1, 10])
ypoints3 = np.array([4, 9, 2, 11])

plt.plot(ypoints, linestyle = 'dotted', color = 'r')
plt.plot(ypoints2, ls = 'dashed', color = 'fuchsia')
plt.plot(ypoints3, ls = '-', color = 'yellowgreen')
plt.show()
```

#### Line Width
- Use the keyword argument linewidth or the shorter lw to change the width of the line.
- The value is a floating number, in points:

```python
#Plot with 3 different widths 5pt, 10pt, 15pt wide line:

import matplotlib.pyplot as plt
import numpy as np


ypoints = np.array([2, 7, 0, 9])
ypoints2 = np.array([3, 8, 1, 10])
ypoints3 = np.array([4, 9, 2, 11])

plt.plot(ypoints, linestyle = 'dotted', color = 'r', linewidth = '5')
plt.plot(ypoints2, ls = 'dashed', color = 'fuchsia', lw = '10')
plt.plot(ypoints3, ls = '-', color = 'yellowgreen', lw = '15')
plt.show()
```

#### Plot as many lines as possible
- Plot many lines by adding the points for the x- and y-axis for each line in the same plt.plot() function.
- If only y-axis values are specified, the x-axis got the the default values (0, 1, 2, 3)
- The x- and y- values come in pairs:

```python
#Draw two lines by specifiyng the x- and y-point values for both lines:

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

#### Matplotlib Labels and Title
- Create Labels for a Plot
- With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

#### Create a Title for a Plot
With Pyplot, you can use the title() function to set a title for the plot.

```python
#Add labels to the x- and y-axis:
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y, color = 'r', ls = '--', ms = '5')

plt.title('Sports Watch Data')
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')

plt.show()
```

#### Set Font Properties for Title and Labels
You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.

```python
# Set font properties for the title and labels:
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'sans serif','color':'blue','size':20}
font2 = {'family':'sans serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```
